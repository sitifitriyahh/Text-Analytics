```python
import pandas as pd

# Load the CSV file
csv_file = "lazada_reviews.csv"  # Ensure this file is in the same directory as the script
df = pd.read_csv(csv_file)

# Inspect the first few rows to find column names
print(df.head())  # Check column names and structure

# Select the required columns (Modify column names if they are different)
columns_needed = ["Reviewer Name", "Review Date", "Review Content"]  # Adjust based on actual column names
df_selected = df[columns_needed]

# Save the extracted data to a new CSV file
output_file = "extracted_reviews.csv"
df_selected.to_csv(output_file, index=False)

print(f"Extracted data saved to {output_file}")
```
